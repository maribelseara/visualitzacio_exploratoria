## - Exercici 1
Descarrega el dataset adjunt preu lloguer per trimestre i barri.csv extret de la web OpenDataBCN i resumeix-lo estadísticament i gràficament.



```python
df=pd.read_csv("preu lloguer per trimestre i barri.csv")
df.head(20)
```

```python
df.info()
```

```python
df.describe()
```

```python
#Treurem la columna Any, ja que tots els valors són 2021 i no ens aporta informació
df=df.drop('Any', axis='columns')
```

```python
df.isna().sum()
```

```python
#Veiem que hi ha registres que tenen l'atribut Preu a -- Hem d'arreglar això per poder fer l'anàlisi
df['Preu'].value_counts()
```

```python
#Cambiem els valors '--' a NaN que reflexa millor la falta de valor
df['Preu']=df['Preu'].replace('--', np.nan)
df['Preu'].isna().sum()
```

```python
#Ara que hem substituit els valors string '--' per NaN podem cambiar el tipus de la columna a un més adient
df['Preu']=df['Preu'].astype('float64')
df.info()
```

```python
#Abans d'ocupar-nos dels NaN dels preus anem a dividir el preus en dos columnes la mensual i la de m2
m2=df[df['Lloguer_mitja']=='Lloguer mitjà per superfície (Euros/m2 mes)'].loc[:,['Trimestre','Codi_Barri','Preu']].reset_index(drop=True)
df=df[df['Lloguer_mitja']=='Lloguer mitjà mensual (Euros/mes)']
#Eliminem la col·lumna Lloguer_mitjà ja que ja no la necesitarem
df=df.drop('Lloguer_mitja', axis='columns')
df=pd.merge(df, m2, on=['Trimestre', 'Codi_Barri'])
df.head(25)
```

```python
#Renombrem les columnes
df=df.rename(columns={ 'Preu_x': 'Preu_mensual', 'Preu_y': 'Preu_m2_mensual'})
df.head(10)
```

```python
#Ara ens ocuparem dels NaN de Preu_Mensual
df[df['Preu_mensual'].isna()].loc[:,['Nom_Barri']].groupby(by='Nom_Barri').value_counts()
```

```python
df.loc[df['Nom_Barri']=='Baró de Viver'].loc[:, ['Nom_Barri', 'Trimestre', 'Preu_mensual']]
```

```python
#Copiarem el preu del segon trimestre al primer i el del quart al tercer per tal de poder fer mitjes anuals
df.loc[df['Nom_Barri']=='Baró de Viver', 'Preu_mensual'] = df.loc[df['Codi_Barri']==58, 'Preu_mensual'].bfill()
df[df['Nom_Barri']=='Baró de Viver'].loc[:,['Nom_Barri', 'Trimestre', 'Preu_mensual']]
```

```python
#Farem el mateix amb el Preu_m2_mensual
df.loc[df['Nom_Barri']=='Baró de Viver'].loc[:, ['Nom_Barri', 'Trimestre', 'Preu_m2_mensual']]
```

```python
df.loc[df['Nom_Barri']=='Baró de Viver', 'Preu_m2_mensual'] = df.loc[df['Codi_Barri']==58, 'Preu_m2_mensual'].bfill()
df.loc[df['Nom_Barri']=='Baró de Viver'].loc[:, ['Nom_Barri', 'Trimestre', 'Preu_m2_mensual']]
```

```python
#A la resta de casos on hem d'omplir els NaN que falten no tenim cap valor de referència del propi barri així que ho haurem de
#fer per interpol·lació
df=df.interpolate()
df.isna().sum()
```

```python
# Com el nom del districte Sarrià-Sant Gervasi és molt llarg afecta a la presentació dels gràfics, l'abreviarem
df['Nom_Districte']=df['Nom_Districte'].replace('Sarrià-Sant Gervasi', 'Sarria-St Gervasi')
```

Ara que tenim les dades netes podem començar a graficar

### Crea almenys una visualització per:

```python
plt.style.use('bmh')
plt.figure(figsize=[14,8])
#Especifiquem les dades del primer trimestre per a que no quadrupliqui el barris
df[df['Trimestre']== 1].loc[:,'Nom_Districte'].value_counts().plot(kind='bar')
plt.title('Quantitat de barris per districte', weight='bold', size=20)
plt.grid(axis='y')
plt.savefig('barris_per_districte.png')
plt.show()
```

En aquest gràfic podem veure la concentració de preus del lloguers a la ciutat de Barcelona, com podem veure al primer gràfic, la majoría dels lloguers es concentren entre els 600 i els 1000€, fins i tot si volguèssim apurar més diriem que entre els 750 i els 850 és on la major part dels lloguers es concentra. Podem veure que la inmensa majoria és concentra entre els 10 i els 14€ per metre quatrat al mes. Particularment entre els 11,5 i els 13,5€

```python
plt.figure(figsize=(16,8))
plt.bar(sorted(df['Nom_Districte'].unique()), df.groupby('Nom_Districte')['Preu_mensual'].agg(np.mean))
plt.title("Mitja de preu de lloguer mensual per districte", size=20, weight='bold')
plt.xlabel("Districtes")
plt.ylabel("Preu mensual(€)")
plt.savefig('Mitja_preu_per_districte.png')
plt.show()
```

En aquest gràfic podem observar la mitja del preu del lloguer mensual per a cada districte. Podem observar així que els districtes amb els lloguers mitjos més cars són Sarrià-Sant Gervasi i Les Corts, mentre que els districtes amb lloguers mitjos més barats són Nou Barris i Sant Andreu.

### Dues variables numèriques (Any o Trimestre i Preu)

```python
plt.figure(figsize=(12,6))
plt.plot(sorted(df['Trimestre'].unique()), df.groupby('Trimestre')['Preu_mensual'].agg(np.mean))
plt.title("Mitja de preu de lloguer mensual per trimestre", size=20, weight='bold')
plt.xlabel("Trimestre")
plt.ylabel("Preu mensual mig(€)")
plt.savefig('Mitja_preu_trimestre.png')
plt.show()
```

A aquest gràfic podem observar la variació del preu mig del lloguer durant els quatre timestres de l'any 2021. Es veu clarament que hi ha haver una baixada del primer al segon trimestre, pero que del segon al tercer trimestre no només és va recuperar sinò que a més va tenir una pujada molt pronunciada. Del tercer al quart trimestre el preu mig del lloguer va seguir pujant, però d'una manera més moderada que al periode anterior.

### Tres variables (Barri o Districte, Trimestre i Preu)

```python
plt.figure(figsize=(14,6))
for element in df['Nom_Districte'].unique():
    plt.plot(sorted(df['Trimestre'].unique()), df[df['Nom_Districte']==element].groupby('Trimestre')['Preu_mensual'].agg(np.mean))
plt.title("Mitja de preu de lloguer mensual per trimestre i districte", size=18, weight='bold')
plt.xlabel("Trimestre")
plt.ylabel("Preu mensual mig(€)")
plt.legend(df['Nom_Districte'].unique(), loc=1)
plt.savefig('Mija_preu_trimestre_districte.png')
plt.show()
```

En aquest gràfic podem veure l'evolució dels preus mitjos de lloguer durant els diferents trimestres de l'any 2021 dividits per districte. Aquí podem veure que la baixada que veiem al gràfic anterior, era causada principalment per una caiguda de preus al districte de Sarrià-Sant Gervasi, que és el que té els preus més alts. En altres barris la baixada ha estat molt menor o fins i tot hi ha hagut certa pujada.
El mateix passa amb la gran pujada que veiem del segon al tercer trimestre, és pronunciada als districtes de Les Corts (el segón més car) i el de Sant Martí, mentre que a la resta de barris la pujada és molt més discreta, gairebé imperceptible en algun d'ells.
Pel periode del tercer al quart trimestre veiem una oposició entre els dos districtes amb lloguers més alts, un puja i un baixa, de manera pronunciada. Això fa que es contraresting entre ells i el gràfic general mostri una pujada lleugera, que és el que observem a la resta de districtes.
Així doncs, podem dir que la situació de preus de lloguer tot i tenir una tendència global pot ser molt diferent per a cadascú dels districtes.

### - Exercici 3
Proposa alguna visualització que creguis que pot resultar interessant.

Hem fet varis gràfics amb el preu mensual però també serà interesant fer servir el preu mensual per m2 ja que el tamany de les cases
pot fer que un barri/districte que sembla molt car en realitat no ho sigui o viceversa, que un barri/districte sembli molt barat però en realitat sigui perquè té vivendes molt petites

```python
plt.figure(figsize=(16,10))
plt.suptitle('Mitjes de preu per districte', size=20, weight='bold')
plt.subplot(2,1,1)
plt.bar(sorted(df['Nom_Districte'].unique()), df.groupby('Nom_Districte')['Preu_mensual'].agg(np.mean))
plt.title("Lloguer mensual per districte", size=15, weight='bold')
plt.xlabel("Districtes")
plt.ylabel("Preu mensual(€)")
plt.subplot(2,1,2)
plt.bar(sorted(df['Nom_Districte'].unique()), df.groupby('Nom_Districte')['Preu_m2_mensual'].agg(np.mean))
plt.title("Lloguer mensual per m2 per districte", size=15, weight='bold')
plt.xlabel("Districtes")
plt.ylabel("Preu mensual m2(€)")
plt.savefig('Preu_mensual_i_m2_districtes.png')
plt.show()
```

En aquesta comparació veiem que quan grafiquem amb el valor de preu mensual per m2 les diferències no són tan pronunciades, en part perquè el rang de valors és més petit i en part perquè si sembla haver-hi diferència en el tamany de les vivendes per districte.
Com que tenim el preu mensual mig i el preu mensual per m2 mig podem calcular quina seria la superfície mitja de les vivendes a cada districte.

```python
plt.figure(figsize=(16,8))
plt.bar(sorted(df['Nom_Districte'].unique()), (df.groupby('Nom_Districte')['Preu_mensual'].agg(np.mean) / df.groupby('Nom_Districte')['Preu_m2_mensual'].agg(np.mean)))
plt.title("Mitja de superfície de les vivendes per districte", size=19, weight='bold')
plt.xlabel("Districtes")
plt.ylabel("Superfície mitja (m2)")
plt.savefig('Superficie_mitja_per_districte.png')
plt.show()
```

Tal com sospitavem els barris amb preu_mitjà de lloguer més car, a part de tenir el preu/m2 més car també tenen vivendes més grans. O al menys algun dels seus barris. El que era el segon més car en preu mensual i preu per m2, queda el primer en quant tamany mitjà de les vivendes.
Volem més informació d'aquest districte en concret, Les Corts,  que a més es el que menys quantitat de barris té i el que ens havia donat una distribució més amplia i més allunyada de la mitja. Grafiquem els seus barris per veure si trobem diferencies significatives entre ells ja sigui en preu o en tamany de les vivendes.

```python
plt.figure(figsize=(15,6))
plt.suptitle('Preus lloguer per barri al districte de Les Corts', size=20, weight='bold')
plt.subplot(1,2,1)
plt.bar(sorted(df.loc[df['Nom_Districte']=='Les Corts', 'Nom_Barri'].unique()), df[df['Nom_Districte']=='Les Corts'].groupby('Nom_Barri')['Preu_mensual'].agg(np.mean))
plt.title("Mitja de preu de lloguer mensual", size=14, weight='bold')
plt.xlabel("Barris")
plt.ylabel("Preu mensual(€)")
plt.subplot(1,2,2)
plt.bar(sorted(df.loc[df['Nom_Districte']=='Les Corts', 'Nom_Barri'].unique()), df[df['Nom_Districte']=='Les Corts'].groupby('Nom_Barri')['Preu_m2_mensual'].agg(np.mean))
plt.title("Mitja de preu de lloguer mensual per m2", size=14, weight='bold')
plt.xlabel("Barris")
plt.ylabel("Preu mensual m2(€)")
plt.savefig('Lloguer_barris_les_Corts.png')
plt.show()
```

Al gràfic anterior podem apreciar que el Barri de Pedralbes clarament està augmentant el preu mig del districte ja que té un preu mig molt superior als altres dos. Quant al preu per m2 també es superior, però no està tan desbalancejat. Així que no podem atribuir que el preu de lloguer al barri de Pedralbes sigui pel tamany de les cases. Podem dir que el preu elevat de la vivenda a Pedralbes és una característica del barri en sí.
